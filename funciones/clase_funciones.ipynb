# Clase Completa: Funciones en Python

Las funciones son bloques de código reutilizables que realizan una tarea específica. Ayudan a que el código sea más modular, legible y fácil de mantener.

## 1. Definición y Estructura Básica

Una función se define con la palabra reservada `def`, seguida del nombre de la función y paréntesis `()`.

* **Nombre:** Debe ser descriptivo.
* **Parámetros:** Variables que recibe la función.
* **Docstring:** Descripción opcional de lo que hace la función.
* **Cuerpo:** El código indentado.
* **Return:** El valor que la función devuelve (si no se especifica, devuelve `None`).

```python
def saludar(nombre):
    """Esta función saluda a la persona que se pase por parámetro."""
    mensaje = f"¡Hola, {nombre}!"
    return mensaje

# Ejemplo de uso
resultado = saludar("Juan")
print(resultado)
```

### 1.1 Profundizando en `return`

La palabra reservada `return` tiene dos funciones principales:
1.  **Finalizar la ejecución:** Cuando Python llega a un `return`, sale de la función inmediatamente.
2.  **Enviar un resultado:** Permite que el valor calculado sea asignado a una variable fuera de la función.

#### ¿Qué sucede si NO usamos `return`?
Si una función no tiene un `return` explícito (o tiene un `return` sin valor), Python devuelve automáticamente `None`. Esto es útil para funciones que realizan acciones (como imprimir o guardar archivos) pero no cálculos.

```python
def funcion_sin_return():
    print("Esta función solo imprime, no devuelve nada.")

resultado = funcion_sin_return()
print(f"Resultado obtenido: {resultado}")  # Imprimirá None

def calcular_area_rectangulo(base, altura):
    return base * altura

#llamamos a la funcion calcular_area pero solo le entregamos los parametros
area = calcular_area_rectangulo(10, 5)
print(f"El área es: {area}") # Aquí 'area' sí guarda el valor 50
```

#### Retorno de Múltiples Valores
Python permite devolver varios valores a la vez separados por comas. Técnicamente, devuelve una **tupla**.

```python
def estadisticas_basicas(lista_numeros):
    minimo = min(lista_numeros) #obtendremos el valor minio de la lista de numeros
    maximo = max(lista_numeros) #obtendremos el valor maximo de la lista de numeros
    promedio = sum(lista_numeros) / len(lista_numeros) #creamos la variable promedio que almacena resultados
    return minimo, maximo, promedio  # Retorno múltiple

mi_lista = [10, 20, 30, 40, 50]
mini, maxi, prom = estadisticas_basicas(mi_lista)# le indicamos que los valores min, max y prom estaran en la lista

print(f"Mín: {mini}, Máx: {maxi}, Promedio: {prom}")
```

## 2. Argumentos y Parámetros

Python ofrece mucha flexibilidad para pasar datos a las funciones:

* **Argumentos posicionales:** Se pasan en el mismo orden de la definición.
* **Valores por defecto:** Permiten llamar a la función con menos argumentos.
* **Args (`*args`):** Para recibir un número arbitrario de argumentos posicionales.
* **Kwargs (`**kwargs`):** Para recibir un número arbitrario de argumentos de palabra clave (diccionarios).

```python
def crear_perfil(nombre, edad, ciudad="Desconocida", *hobbies, **detalles_extra):
    print(f"Usuario: {nombre}, Edad: {edad}, Ciudad: {ciudad}")
    print(f"Hobbies: {hobbies}")
    print(f"Extra: {detalles_extra}")

crear_perfil("Ana", 25, "Madrid", "Pintura", "Ciclismo", ocupacion="Ingeniera", premium=True)
```

```python
# Modificamos para permitir el ingreso de detalles extra (diccionario)

def crear_perfil(nombre, edad, ciudad="", *hobbies, **detalles_extra):
    print(f"\n--- Perfil Creado ---")
    print(f"Usuario: {nombre}, Edad: {edad}, Ciudad: {ciudad}")
    print(f"Hobbies: {hobbies}")
    print(f"Extra: {detalles_extra}")

nombre = input("Ingrese su nombre: ")
edad = int(input("Ingrese su edad: "))
ciudad = input("Ingrese su ciudad: ")
hobbies = input("Ingrese sus hobbies separados por comas: ").split(",")

# Para llenar el diccionario **kwargs vía input:
detalles_extra = {}
print("\nIngrese detalles extra (ej: (key)ocupacion y (valor), profesion y la profesion). Escriba 'salir' en nombre para terminar.")

#el for nos hara llenar el diccionario (key y valor)

#el while nos hara el bucle hasta salir, pidiendo los datos del diccionario hasta q indiquemos salir
while True:
    clave = input("Nombre del detalle (o 'salir'): ")
    if clave.lower() == 'salir':
        break
    valor = input(f"Valor para {clave}: ")
    detalles_extra[clave] = valor

# Llamamos a la función expandiendo la lista (*) y el diccionario (**)
crear_perfil(nombre, edad, ciudad, *hobbies, **detalles_extra)
```

## 3. Funciones de Orden Superior y Lambda

En Python, las funciones son objetos de "primera clase", lo que significa que puedes pasarlas como argumentos o devolverlas.

* **Lambda:** Funciones anónimas y rápidas de una sola línea.

```python
# 1. Definición de una función Lambda
# 'cuadrado' es el nombre de la variable que almacenará la función
# 'lambda' es la palabra reservada
# 'x' es el argumento (parámetro de entrada)
# ':' separa el argumento de la expresión
# 'x ** 2' es el cuerpo de la función (lo que retorna automáticamente)
cuadrado = lambda x: x ** 2

print(f"El cuadrado de 5 es: {cuadrado(5)}")

# 2. Uso de Lambda como argumento (Funciones de Orden Superior)
def aplicar_operacion(n, func):
    """Esta función recibe un número 'n' y una función 'func'"""
    return func(n) # Ejecuta la función que le enviamos pasando 'n'

# Aquí enviamos el número 10 y una lambda que multiplica por 3
# El paso a paso es:
# a) Se define la lambda (x: x * 3) en el momento
# b) 'aplicar_operacion' recibe 10 y esa lambda
# c) Retorna 10 * 3
print(f"Aplicando lambda: {aplicar_operacion(10, lambda x: x * 3)}")
```

## 4. Conceptos Avanzados: Getters, Setters y @property

Aunque los getters y setters son comunes en POO (Programación Orientada a Objetos), en Python se manejan de forma elegante mediante decoradores de funciones para controlar el acceso a los atributos.

### Profundizando en Getters, Setters y `@property`

En Python, no usamos métodos como `get_saldo()` o `set_saldo()` como en otros lenguajes. Usamos **decoradores** para que los métodos se comporten como atributos.

*   **`@property` (El Getter):** Se coloca encima de un método para que podamos acceder a él sin usar paréntesis. Su función es **entregar** el valor del atributo interno.
*   **`@atributo.setter` (El Setter):** Se usa para **modificar** el valor. La gran ventaja es que aquí podemos poner filtros (por ejemplo, que la edad no sea negativa o que el saldo sea un número).
*   **El Atributo Interno (`_variable`):** Usamos un guion bajo al principio para indicar que esa variable es 'privada' y que solo debe ser accedida a través del getter y el setter.

## 4.1. Conceptos Clave: `self` y Manejo de Errores (`try...except`)

### ¿Qué es `self`?
En Python, `self` representa la **instancia** de la clase.
* Cuando defines `self.nombre = nombre`, le estás diciendo a Python: 'guarda este nombre específicamente para este objeto'.
* Sin `self`, las variables serían locales a la función y desaparecerían al terminar; con `self`, los datos persisten mientras el objeto exista.

### ¿Qué es `try...except`?
Es una estructura de **control de excepciones**. Sirve para evitar que el programa se cierre abruptamente cuando ocurre un error.

* **`try`**: Aquí pones el código que *podría* fallar (por ejemplo, convertir un texto del usuario a número).
* **`except`**: Aquí defines qué hacer si ocurre el error.
* **`ValueError`**: Es el tipo de error específico que ocurre cuando intentas convertir algo como "hola" en un **texto en negrita**número entero (`int`).

```python
class CuentaBancaria:
    def __init__(self, saldo_inicial):
        self._saldo = saldo_inicial # Atributo privado por convención

    @property
    def saldo():
        """Este es el 'Getter'. Permite leer el valor como un atributo."""
        print("Consultando saldo...")
        return self._saldo

    @saldo.setter
    def saldo(self, nuevo_monto):
        """Este es el 'Setter'. Permite validar antes de cambiar el valor."""
        if nuevo_monto < 0:
            print("Error: El saldo no puede ser negativo.")
        else:
            self._saldo = nuevo_monto

cuenta = CuentaBancaria(100)
print(cuenta.saldo) # Llama al getter
cuenta.saldo = 250   # Llama al setter
cuenta.saldo = -50   # Intento inválido
```

### Ejercicio Práctico: Validación de Edad con Setters

En este ejercicio, crearemos una clase `Persona` donde usaremos un setter para asegurarnos de que la edad ingresada por el usuario sea coherente (no puede ser menor a 0).

```python
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre
        self._edad = 0 # Atributo privado inicial

    @property
    def edad(self):
        """Getter: Retorna la edad actual"""
        return self._edad

    @edad.setter
    def edad(self, valor):
        """Setter: Valida que la edad no sea negativa antes de asignar"""
        if valor < 0:
            print(f"¡Error! {valor} no es una edad válida. La edad se mantendrá en {self._edad}.")
        else:
            print(f"Edad actualizada correctamente a {valor}.")
            self._edad = valor

# --- Uso del programa con Input ---
usuario_nombre = input("Ingrese el nombre de la persona: ")
persone_obj = Persona(usuario_nombre)

try:
    nueva_edad = int(input(f"¿Qué edad tiene {usuario_nombre}?: "))
    # Aquí se activa automáticamente el @edad.setter
    persone_obj.edad = nueva_edad

    print(f"\nResumen: {persone_obj.nombre} tiene {persone_obj.edad} años.")
except ValueError:
    print("Por favor, ingrese un número entero válido para la edad.")
```

```python
def ejemplo_try_self():
    try:
        # Intentamos convertir la entrada a número
        numero = int(input("Ingresa un número para dividir 100: "))
        resultado = 100 / numero
        print(f"El resultado es {resultado}")
    except ValueError:
        print("Error: ¡Debes ingresar un número, no letras!")
    except ZeroDivisionError:
        print("Error: No se puede dividir por cero.")

ejemplo_try_self()
```

### Ejercicio Práctico: Manejo de Errores en Listas

En este ejemplo, manejaremos dos posibles errores comunes al trabajar con entradas del usuario y listas: `ValueError` (si no ingresa un número) e `IndexError` (si el número está fuera de rango).

```python
def elegir_fruta():
    frutas = ["Manzana", "Banana", "Cereza", "Durazno"]
    print(f"Lista de frutas: {frutas}")

    try:
        # Pedimos el índice (posición) de la fruta
        indice = int(input("Ingresa el índice de la fruta que quieres (0 a 3): "))

        # Intentamos acceder al elemento de la lista
        seleccion = frutas[indice]
        print(f"Has seleccionado: {seleccion}")

    except ValueError:
        print("Error: Por favor, ingresa solo números enteros.")
    except IndexError:
        print(f"Error: El índice debe estar entre 0 y {len(frutas)-1}. No existe esa posición.")
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")

elegir_fruta()
```

### Combinando `while` y `try...except` para Validaciones Robustas

Esta es la forma más profesional de pedir datos al usuario. El bucle `while` se repetirá infinitamente hasta que el código llegue a la instrucción `break`, la cual solo se ejecutará si no ocurrió ninguna excepción en las líneas anteriores.

```python
def pedir_numero_valido():
    while True:
        try:
            # Intentamos obtener y convertir el dato
            valor = int(input("Por favor, ingresa un número entero positivo: "))

            if valor < 0:
                print("El número debe ser positivo. Intenta de nuevo.")
                continue # Regresa al inicio del while

            # Si llegamos aquí, el dato es correcto y salimos del bucle
            return valor

        except ValueError:
            # Si el usuario ingresa letras, el programa viene aquí en lugar de cerrarse
            print("¡Error! Eso no es un número entero. Intenta de nuevo.")

# Uso de la función
numero_final = pedir_numero_valido()
print(f"Gracias, el número aceptado fue: {numero_final}")
```

## 5. Decoradores

Un decorador es una función que envuelve a otra función para extender su comportamiento sin modificar su estructura original.

```python
def mi_decorador(func):
    def envoltura():
        print("--- Antes de ejecutar la función ---")
        func()
        print("--- Después de ejecutar la función ---")
    return envoltura

@mi_decorador
def tarea_especifica():
    print("Ejecutando la tarea central...")

tarea_especifica()
```

### Ejercicio Práctico: Decorador con Input

Vamos a crear un decorador llamado `decorador_estetico` que añade bordes a cualquier mensaje que el usuario decida escribir.

```python
def decorador_estetico(func):
    def envoltura(texto):
        print("\n" + "*" * (len(texto) + 4))
        func(texto)
        print("*" * (len(texto) + 4) + "\n")
    return envoltura

@decorador_estetico
def mostrar_mensaje(m):
    print(f"* {m} *")

# Pedimos el dato al usuario
usuario_texto = input("Escribe una frase para decorar: ")

# Llamamos a la función
mostrar_mensaje(usuario_texto)
```

### Ejercicio Avanzado: Decorador de Seguridad con Bucles

En este escenario, crearemos un decorador que valida si un usuario tiene permiso para ejecutar una función sensible (como `ver_reporte_ventas`). El sistema dará 3 intentos usando un bucle `while` y buscará al usuario en una lista permitida usando `for`.

```python
def validar_acceso(func):
    def envoltura(*args, **kwargs):
        usuarios_autorizados = ["admin", "gerente", "supervisor"]
        intentos_maximos = 3
        intento_actual = 0

        print(f"--- Sistema de Seguridad --- (Intentos permitidos: {intentos_maximos})")

        while intento_actual < intentos_maximos:
            usuario = input("Ingrese su nombre de usuario (admin, gerente o supervisor): ").lower()
            encontrado = False

            # Usamos for para buscar en la lista de autorizados
            for u in usuarios_autorizados:
                if u == usuario:
                    encontrado = True
                    break

            if encontrado:
                print(f"¡Acceso concedido para {usuario}!")
                return func(*args, **kwargs) # Ejecuta la función original
            else:
                intento_actual += 1
                print(f"Usuario incorrecto. Te quedan {intentos_maximos - intento_actual} intentos.")

        print("\n[BLOQUEO] Se han agotado los intentos. Contacte al administrador.")
    return envoltura

@validar_acceso
def ver_reporte_ventas():
    print("\n--- REPORTE DE VENTAS ---")
    print("Enero: $10,000 | Febrero: $12,500 | Marzo: $9,800")
    print("-------------------------")

# Ejecución del ejercicio
ver_reporte_ventas()
```

## 🏆 Desafío Final: Sistema de Inventario Protegido

Este ejercicio integra casi todo lo aprendido: diccionarios, bucles `while`, entrada de datos, manejo de errores y decoradores de seguridad.

```python
def requerir_password(password_correcta):
    def decorador(func):
        def envoltura(*args, **kwargs):
            intento = input("\n[SEGURIDAD] Ingrese la contraseña de administrador para ver el reporte: ")
            if intento == password_correcta:
                return func(*args, **kwargs)
            else:
                print("Acceso denegado. Contraseña incorrecta.")
        return envoltura
    return decorador

# --- Configuración inicial ---
print("--- Configuración del Sistema ---")
admin_pass = input("Cree una contraseña para el sistema: ")
inventario = {}

# --- Carga de datos ---
print("\n--- Registro de Productos (Escriba 'fin' en el producto para terminar) ---")
while True:
    producto = input("Nombre del producto: ")
    if producto.lower() == 'fin':
        break
    try:
        precio = float(input(f"Precio de {producto}: "))
        inventario[producto] = precio
    except ValueError:
        print("Error: Ingrese un valor numérico para el precio.")

# --- Funcion del Reporte ---
@requerir_password(admin_pass)
def generar_reporte(datos):
    print("\n" + "="*30)
    print("REPORTE DE INVENTARIO")
    print("="*30)

    total_productos = len(datos)
    suma_precios = sum(datos.values())

    for prod, precio in datos.items():
        print(f"• {prod.capitalize()}: ${precio:,.2f}")

    print("-"*30)
    print(f"Stock Total: {total_productos} items")
    print(f"Valor Total del Inventario: ${suma_precios:,.2f}")
    print("="*30)

# --- Ejecución ---
generar_reporte(inventario)
```
