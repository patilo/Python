# Clase: Control de Flujo con el Bucle `while` en Python

## 1. Definición y Estructura
El bucle `while` permite ejecutar un bloque de código **mientras** una condición sea verdadera (`True`). Es ideal cuando no sabemos de antemano cuántas veces necesitaremos repetir una acción.

### Estructura básica:
```python
while condicion:
    # Bloque de código a repetir
```

## 2. Diferencias Clave

| Estructura | Función Principal |
| :--- | :--- |
| **`if`** | Ejecuta un bloque **una sola vez** si la condición es verdadera. |
| **`match-case`** | Compara un valor contra varios patrones (más eficiente y limpio que muchos `if-elif`). |
| **`while`** | Crea un ciclo que **repite** el bloque hasta que la condición sea falsa. |

## 3. Control del Bucle: `break`, `continue` y `else`

*   **`break`**: Sale inmediatamente del bucle.
*   **`continue`**: Salta el resto del código en la iteración actual y vuelve a evaluar la condición.
*   **`else`**: Ejecuta un bloque de código cuando el bucle termina normalmente (sin ser interrumpido por un `break`).

````python
# === EJEMPLO BÁSICO: Contador simple ===
print("--- Ejemplo Básico ---")
contador = 1
while contador <= 3:
    print(f"Contador: {contador}")
    contador += 1 # Importante: incrementar para evitar bucle infinito
````

````python
# === EJEMPLO INTERMEDIO: Menú Interactivo con Break/Continue ===
print("\n--- Ejemplo Intermedio ---")
while True:
    opcion = input("Escribe 'hola', 'saltar' o 'salir': ").lower()

    if opcion == 'salir':
        print("Saliendo del programa...")
        break # Rompe el bucle

    if opcion == 'saltar':
        print("Saltando esta parte...")
        continue # Vuelve al inicio del while

    print(f"Escribiste: {opcion}")
````

````python
# === EJEMPLO AVANZADO: Validación de datos y Clausula Else ===
print("\n--- Ejemplo Avanzado ---")
intentos = 0
max_intentos = 3

while intentos < max_intentos:
    clave = input(f"Intento {intentos+1}: Ingresa la clave secreta: ")
    if clave == "python123":
        print("¡Acceso concedido!")
        break
    print("Clave incorrecta.")
    intentos += 1
else:
    # Este bloque solo corre si el while NO terminó por un 'break'
    print("Has agotado tus 3 intentos. Cuenta bloqueada.")
````

## Conceptos de Contador y Acumulador

En programacion competitiva y logica de negocios, estos dos conceptos son fundamentales:
*   **Contador:** Una variable que se incrementa en una cantidad fija (ej. `c = c + 1`). Sirve para saber *cuantas veces* ocurre algo.
*   **Acumulador:** Una variable que se incrementa en una cantidad variable (ej. `a = a + valor`). Sirve para obtener un *total* o *suma*.

````python
# === EJERCICIO: Contador y Acumulador ===
print("--- Registro de Ventas ---")

total_dinero = 0  # Este es el ACUMULADOR
cantidad_ventas = 0  # Este es el CONTADOR

while True:
    monto = float(input("Ingrese el monto de la venta (0 para finalizar): "))

    if monto == 0:
        break

    if monto < 0:
        print("Monto invalido, intente de nuevo.")
        continue

    total_dinero += monto      # Acumulamos el valor
    cantidad_ventas += 1       # Contamos la transaccion

print(f"\nResumen del d3a:")
print(f"Ventas realizadas: {cantidad_ventas}")
print(f"Total recaudado: ${total_dinero:.2f}")
````

## 4. Ejercicio Resuelto
**Enunciado:** Crear un programa que sume números ingresados por el usuario hasta que ingrese un número negativo. No debe sumar el número negativo.

```python
suma = 0
while True:
    num = int(input("Ingresa un número para sumar (negativo para terminar): "))
    if num < 0:
        break # Salimos sin sumar
    suma += num
print(f"La suma total es: {suma}")
```
