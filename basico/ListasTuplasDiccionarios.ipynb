# Estructuras de Datos en Python

En esta sesión aprenderemos las tres estructuras de datos fundamentales en Python:
1. **Listas**: Mutables y ordenadas.
2. **Tuplas**: Inmutables y ordenadas.
3. **Diccionarios**: Mutables y basados en pares clave-valor.

## Bonus: Manejo de Excepciones (try-except)

Antes de entrar en estructuras de datos, es vital saber manejar errores.

**¿Qué es?** Es una estructura que permite que tu programa no se 'rompa' (detenga) cuando ocurre un error durante la ejecución.

**¿Cuándo usarlo?**
*   Cuando pides datos al usuario (input).
*   Cuando realizas operaciones matemáticas que podrían fallar (división por cero).
*   Cuando accedes a archivos o bases de datos.

```python
# --- EJEMPLO RESUELTO: try-except ---
# Tarea: Pedir un número al usuario y dividir 100 por ese número.

try:
    # 1. Intentamos realizar el código que podría fallar
    valor = input("Introduce un número para dividir 100: ")
    numero = float(valor)
    resultado = 100 / numero

except ValueError:
    # 2. Capturamos el error si el usuario no ingresó un número
    print("ERROR: Debes ingresar un número válido (caracteres no permitidos).")

except ZeroDivisionError:
    # 3. Capturamos el error si el número es cero
    print("ERROR: No se puede dividir por cero.")

except Exception as e:
    # 4. Capturamos cualquier otro error inesperado
    print(f"Ocurrió un error inesperado: {e}")

else:
    # 5. Se ejecuta solo si NO hubo errores
    print(f"¡Éxito! El resultado es: {resultado}")

finally:
    # 6. Se ejecuta SIEMPRE (haya error o no)
    print("Operación finalizada.")
```

## Comparativa: while vs match-case

Es común confundir estructuras de control, pero tienen propósitos muy distintos:

### 1. Bucle `while` (Repetición)
Se usa cuando quieres ejecutar un bloque de código **muchas veces** mientras una condición sea verdadera.
*   **Ejemplo:** 'Mientras el usuario no escriba salir, sigue pidiendo datos'.

### 2. `match case` (Selección/Patrones)
Se usa para **elegir un camino** entre varias opciones dependiendo del valor de una variable. Es una alternativa más limpia y potente a múltiples `if-elif-else`.
*   **Ejemplo:** 'Si el comando es 1, suma; si es 2, resta; si es cualquier otra cosa, da error'.

```python
# --- EJEMPLO COMPARATIVO ---

print("--- Ejemplo de match-case ---")
dia = "Lunes"

match dia:
    case "Lunes":
        print("Inicio de semana.")
    case "Viernes":
        print("Casi fin de semana.")
    case _:
        print("Día normal.")

print("\n--- Ejemplo de while ---")
contador = 3
while contador > 0:
    print(f"Cuenta regresiva: {contador}")
    contador -= 1
print("¡Despegue!")
```

## 1. Listas (Lists)
Las listas son colecciones de elementos que pueden ser modificados (mutables).

```python
# --- CRUD y Operaciones en Listas ---

# Crear (Create)
frutas = ["manzana", "pera", "banana"]

# Añadir elementos
frutas.append("uva") # si no indicamos la posición, esta sera agregada al final
frutas.insert(1, "naranja") # En una posición específica

# Leer (Read)
print(f"Primera fruta: {frutas[0]}")
print(f"Cantidad de elementos (len): {len(frutas)}")

# Buscar (Search)
existe = "pera" in frutas
indice = frutas.index("banana")

# Actualizar / Cambiar (Update)
frutas[2] = "kiwi"

# Eliminar (Delete)
frutas.remove("manzana") # Por valor
eliminado = frutas.pop(0) # Por índice y retorna el valor

# --- Propiedades Aritméticas ---
numeros = [10, 20, 30, 40, 50]
print(f"Suma: {sum(numeros)}")
print(f"Máximo: {max(numeros)}")
print(f"Mínimo: {min(numeros)}")
print(f"Promedio: {sum(numeros)/len(numeros)}")

# --- Creación con Bucles ---
cuadrados = []
for x in range(5):
    cuadrados.append(x**2)

# List Comprehension (Alternativa pro)
pares = [x for x in range(10) if x % 2 == 0]

display("Listas finales:", frutas, numeros, cuadrados, pares)
```

## 2. Tuplas (Tuples)
Las tuplas son similares a las listas pero **no se pueden cambiar** después de su creación (son inmutables).

```python
# --- Operaciones en Tuplas ---

# Crear
mi_tupla = (1, 5, 8, 5, 2)

# Leer
print(f"Elemento en índice 2: {mi_tupla[2]}")

# Buscar / Contar
print(f"¿Cuántas veces aparece el 5?: {mi_tupla.count(5)}")
print(f"Índice del 8: {mi_tupla.index(8)}")

# Nota sobre CRUD:
# No puedes hacer .append(), .remove() o cambiar mi_tupla[0] = X.
# Para 'modificar', debes convertir a lista y luego volver a tupla:
temp_list = list(mi_tupla)
temp_list.append(100)
mi_tupla_nueva = tuple(temp_list)

# Propiedades Aritméticas (Igual que listas)
print(f"Suma de tupla: {sum(mi_tupla)}")

# Creación con bucles (usando el constructor tuple)
tupla_while = []
i = 0
while i < 3:
    tupla_while.append(i)
    i += 1
tupla_while = tuple(tupla_while)

display(mi_tupla, mi_tupla_nueva, tupla_while)
```

## 3. Diccionarios (Dictionaries)
Estructuras de clave:valor. Ideales para representar objetos con propiedades.

```python
# --- CRUD y Operaciones en Diccionarios ---

# Crear
estudiante = {
    "nombre": "Juan",
    "edad": 21,
    "carrera": "Ingeniería"
}

# Leer
print(f"Nombre: {estudiante['nombre']}")
print(f"Usando get (seguro): {estudiante.get('nota', 'No asignada')}")

# Actualizar y Añadir
estudiante["edad"] = 22 # Cambiar existente
estudiante["ciudad"] = "Madrid" # Añadir nuevo

# Eliminar
del estudiante["carrera"]
eliminado_val = estudiante.pop("ciudad")

# --- Llenar datos con .items() ---
# Supongamos que queremos copiar o transformar un diccionario
nuevo_dicc = {}
for clave, valor in estudiante.items():# como ven, indicamos el diccionario.items para el llenado
    nuevo_dicc[clave] = f"Dato: {valor}"

# Operaciones Aritméticas en Diccionarios (aplicado a valores)
notas = {"Mate": 90, "Fisica": 85, "Programacion": 100}
print(f"Nota máxima: {max(notas.values())}")
print(f"Promedio de notas: {sum(notas.values())/len(notas)}")

display(estudiante, nuevo_dicc)
```

## Ejemplos Resueltos Paso a Paso

```python
# --- EJEMPLO RESUELTO: LISTAS ---
# Tarea: Filtrar una lista de edades para obtener solo los mayores de edad y calcular su promedio.

edades_crudas = [12, 18, 25, 14, 30, 10, 45]
edades_filtradas = []

# 1. Recorremos con un bucle for
for edad in edades_crudas:
    # 2. Buscamos/Validamos una condición
    if edad >= 18:
        # 3. Añadimos (Update/Create) a la nueva lista
        edades_filtradas.append(edad)

# 4. Operaciones aritméticas
if len(edades_filtradas) > 0:
    suma_edades = sum(edades_filtradas)
    promedio = suma_edades / len(edades_filtradas)
    print(f"Mayores de edad: {edades_filtradas}")
    print(f"Promedio de edad legal: {promedio:.2f}")
```

```python
# --- EJEMPLO RESUELTO: DICCIONARIOS ---
# Tarea: Gestionar un carrito de compras y aplicar un 10% de descuento a cada item.

carrito = {
    "Laptop": 1200,
    "Raton": 25,
    "Teclado": 45
}

carrito_descuento = {}

# 1. Usamos .items() para obtener clave y valor simultáneamente
for producto, precio in carrito.items():
    # 2. Calculamos el nuevo precio (Operación aritmética)
    precio_final = precio * 0.90
    # 3. Llenamos el nuevo diccionario
    carrito_descuento[producto] = precio_final

print("Precios originales:", carrito)
print("Precios con 10% desc:", carrito_descuento)

# 4. Buscar un producto específico de forma segura
busqueda = "Monitor"
resultado = carrito_descuento.get(busqueda, "El producto no está en el carrito")
print(f"Resultado de búsqueda ({busqueda}): {resultado}")
```
