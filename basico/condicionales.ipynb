# Ejercicios de Condicionales en Python

## Introducción a las Sentencias Condicionales: `if`, `elif`, `else`

Las sentencias condicionales nos permiten ejecutar diferentes bloques de código dependiendo de si una condición es verdadera o falsa. Son como tomar decisiones en la vida real: "Si llueve, llevo paraguas; si no, no lo llevo."

*   **`if`**: La condición principal. Si es verdadera, se ejecuta el código dentro de este bloque.
*   **`elif`** (abreviatura de "else if"): Una condición alternativa. Si la condición del `if` anterior fue falsa, se evalúa esta. Puedes tener múltiples `elif`.
*   **`else`**: El bloque final. Si todas las condiciones `if` y `elif` anteriores fueron falsas, se ejecuta el código dentro del `else`.

**Recuerda la sintaxis:**
```python
if condicion1:
    # Código si condicion1 es Verdadera
elif condicion2:
    # Código si condicion1 es Falsa Y condicion2 es Verdadera
else:
    # Código si todas las condiciones anteriores son Falsas
```

````python
# Ejercicio 1: Determinar si un número es positivo, negativo o cero

# Pedimos al usuario que ingrese un número
numero_ingresado = float(input("Por favor, ingresa un número: "))

# Usamos la sentencia if-elif-else para evaluar el número
if numero_ingresado > 0:
    # Si el número es mayor que 0, es positivo
    print(f"El número {numero_ingresado} es POSITIVO.")
elif numero_ingresado < 0:
    # Si la condición anterior (numero_ingresado > 0) es falsa
    # y el número es menor que 0, entonces es negativo
    print(f"El número {numero_ingresado} es NEGATIVO.")
else:
    # Si ninguna de las condiciones anteriores es verdadera (no es mayor ni menor que 0),
    # entonces el número debe ser cero
    print(f"El número {numero_ingresado} es CERO.")
````

## Anidación de Sentencias Condicionales

La anidación significa colocar una sentencia condicional dentro de otra. Esto es útil cuando necesitamos tomar decisiones más complejas, donde la segunda decisión depende del resultado de la primera. Imagina: "Si tienes hambre, entonces: si tienes dinero, compra comida; si no, pide prestado."

**Sintaxis:**
```python
if condicion_exterior:
    # Código si condicion_exterior es Verdadera
    if condicion_interior:
        # Código si ambas condiciones son Verdaderas
    else:
        # Código si condicion_exterior es Verdadera Y condicion_interior es Falsa
else:
    # Código si condicion_exterior es Falsa
```

````python
# Ejercicio 2: Determinar si un estudiante es apto para un curso avanzado
# Un estudiante es apto si tiene más de 18 años Y (una nota mayor a 80 O un certificado especial).

# Pedimos al usuario la edad y la nota
edad_estudiante = int(input("Ingresa la edad del estudiante: "))
nota_estudiante = int(input("Ingresa la nota del estudiante (0-100): "))
tiene_certificado = input("¿Tiene un certificado especial? (si/no): ").lower() # Convertimos a minúsculas para facilitar la comparación

# Usamos anidación para evaluar múltiples condiciones
if edad_estudiante >= 18:
    # Primera condición: el estudiante es mayor o igual a 18 años
    print("El estudiante tiene la edad requerida.")

    if nota_estudiante > 80 or tiene_certificado == 'si':
        # Segunda condición (anidada): el estudiante tiene buena nota O un certificado
        # Utilizamos 'or' porque solo se necesita que una de las dos sea verdadera
        # Utilizamos '==' para comparar strings (si el string es exactamente 'si')
        print("El estudiante cumple con la nota o tiene el certificado.")
        print("¡Felicidades! El estudiante es APTO para el curso avanzado.")
    else:
        # Si la edad es correcta, pero no cumple ni la nota ni el certificado
        print("El estudiante no cumple con la nota o el certificado.")
        print("El estudiante NO es apto para el curso avanzado (le faltan requisitos de nota/certificado).")
else:
    # Si la primera condición (edad) no se cumple
    print("El estudiante no tiene la edad requerida.")
    print("El estudiante NO es apto para el curso avanzado (por edad).")
````

## Comparación de Strings (`==`) y Números (`==`, `!=`, `<`, `>`, `<=`, `>=`)

Es fundamental saber cómo comparar diferentes tipos de datos. No es lo mismo comparar texto que comparar cantidades.

### Comparación de Strings (`==`)

Para comparar si dos textos (strings) son exactamente iguales, usamos el operador de igualdad `==`.

*   `nombre1 == nombre2`: Devuelve `True` si los strings `nombre1` y `nombre2` tienen exactamente los mismos caracteres, en el mismo orden y con la misma capitalización (mayúsculas/minúsculas). De lo contrario, devuelve `False`.

### Comparación de Números (`==`, `!=`, `<`, `>`, `<=`, `>=`)

Para números, tenemos más operadores, ya que podemos verificar igualdad, desigualdad o magnitud:

*   `==` (igual a): `a == b` es `True` si `a` y `b` tienen el mismo valor.
*   `!=` (distinto de): `a != b` es `True` si `a` y `b` tienen valores diferentes.
*   `<` (menor que): `a < b` es `True` si `a` es menor que `b`.
*   `>` (mayor que): `a > b` es `True` si `a` es mayor que `b`.
*   `<=` (menor o igual que): `a <= b` es `True` si `a` es menor o igual que `b`.
*   `>=` (mayor o igual que): `a >= b` es `True` si `a` es mayor o igual que `b`.


````python
# Ejercicio 3: Comparar nombres y edades

# Pedimos dos nombres y dos edades
nombre1 = input("Ingresa el primer nombre: ")
nombre2 = input("Ingresa el segundo nombre: ")
edad1 = int(input("Ingresa la edad de " + nombre1 + ": "))
edad2 = int(input("Ingresa la edad de " + nombre2 + ": "))

# Comparamos los nombres (strings) usando ==
print("\n--- Comparación de Nombres ---")
if nombre1 == nombre2:
    # Si ambos strings son idénticos (incluyendo mayúsculas/minúsculas)
    print(f"¡Ambos se llaman {nombre1}!")
else:
    # Si los strings son diferentes
    print(f"{nombre1} y {nombre2} son nombres diferentes.")

# Comparamos las edades (números) usando varios operadores
print("\n--- Comparación de Edades ---")

if edad1 == edad2:
    # Verificamos si las edades son exactamente iguales
    print(f"{nombre1} y {nombre2} tienen la misma edad: {edad1} años.")
elif edad1 != edad2:
    # Verificamos si las edades son diferentes
    print(f"{nombre1} y {nombre2} tienen edades diferentes.")

if edad1 > edad2:
    # Verificamos si la edad1 es mayor que la edad2
    print(f"{nombre1} es mayor que {nombre2}.")
elif edad1 < edad2:
    # Verificamos si la edad1 es menor que la edad2
    print(f"{nombre1} es menor que {nombre2}.")

# Un ejemplo con <= y >=
if edad1 >= 18:
    # Verificamos si edad1 es mayor o igual a 18
    print(f"{nombre1} tiene {edad1} años y es mayor de edad (>=18).")
else:
    print(f"{nombre1} tiene {edad1} años y es menor de edad (<18).")
````

## Operadores Lógicos: `and` y `or`

Estos operadores nos permiten evaluar varias condiciones en una sola línea.

*   **`and` (Y)**: La condición total es verdadera **solo si todas** las partes son verdaderas.
*   **`or` (O)**: La condición total es verdadera **si al menos una** de las partes es verdadera.

````python
# Ejercicio: Sistema de Acceso a una Atracción
# Para subir, el usuario debe medir más de 140cm Y tener más de 10 años.
# O bien, si es un 'Invitado VIP', puede entrar sin importar lo anterior.

estatura = int(input("Ingresa tu estatura en cm: "))
edad = int(input("Ingresa tu edad: "))
es_vip = input("¿Eres invitado VIP? (si/no): ").lower() == "si"

# Usamos 'or' para el beneficio VIP y 'and' para los requisitos normales
if es_vip or (estatura > 140 and edad > 10):
    # Si es VIP (True), no mira el resto.
    # Si no es VIP, obliga a que estatura Y edad cumplan el mínimo.
    print("¡Acceso concedido! Disfruta la atracción.")
else:
    print("Lo sentimos, no cumples los requisitos de seguridad para subir.")
````

## La Sentencia `match case` (disponible desde Python 3.10+)

La sentencia `match case` es una forma más moderna y elegante de manejar múltiples condiciones, especialmente cuando se basan en el valor exacto de una variable. Es una alternativa a usar muchos `elif` seguidos para comparar el mismo valor. Es muy útil para hacer "coincidencia de patrones" (pattern matching).

**Sintaxis básica:**
```python
variable = valor

match variable:
    case patron1:
        # Código si variable coincide con patron1
    case patron2:
        # Código si variable coincide con patron2
    case _:
        # Código si no coincide con ningún patrón anterior (similar a un 'else')
```

El `_` (guion bajo) actúa como un comodín o "captura todo", similar al `else` en `if/elif/else`.

````python
# Ejercicio 4: Elegir una acción basada en una opción de menú (string)

# Pedimos al usuario que elija una opción
opcion_elegida = input("Elige una opción (A, B, C): ").upper() # Convertimos a mayúsculas para simplificar la comparación

# Usamos match case para ejecutar código diferente según la opción
match opcion_elegida:
    case 'A':
        # Si opcion_elegida es exactamente 'A'
        print("Has elegido la Opción A: ¡Iniciar juego!")
    case 'B':
        # Si opcion_elegida es exactamente 'B'
        print("Has elegido la Opción B: Cargar partida.")
    case 'C':
        # Si opcion_elegida es exactamente 'C'
        print("Has elegido la Opción C: Ver ajustes.")
    case _:
        # Si opcion_elegida no coincide con 'A', 'B' ni 'C'
        print("Opción no válida. Por favor, elige A, B o C.")
````

## `match case` con Múltiples Valores y Anidación Implícita (más avanzados)

`match case` puede hacer más que solo comparar valores exactos. Puede trabajar con tuplas, listas, diccionarios y aplicar condiciones más complejas. Esto lo hace muy potente para "anidación" de patrones.

*   **Múltiples valores para un caso**: Puedes usar `|` (OR lógico) para que un caso coincida con varios valores.
*   **Patrones con variables**: Puedes asignar partes del patrón a variables.
*   **Patrones anidados**: Puedes verificar la estructura de datos más complejas.

**Ejemplo de sintaxis avanzada:**
```python
# Coincidencia con múltiples valores
match dia_semana:
    case 'Sabado' | 'Domingo':
        print("Es fin de semana")

# Coincidencia con tuplas y asignación a variables
match coordenada:
    case (x, y):
        print(f"X es {x}, Y es {y}")

# Coincidencia con diccionarios y condiciones (guard clauses)
match comando:
    case {'accion': 'mover', 'direccion': dir, 'pasos': p} if p > 0:
        print(f"Moviendo {dir} {p} pasos")
    case {'accion': 'parar'}:
        print("Deteniendo")
```

````python
# Ejercicio 5: Clasificar un día de la semana y estado del tiempo (ejemplo con tuplas y múltiples valores)

# Pedimos el día de la semana y el estado del tiempo
dia = input("Ingresa el día de la semana (Lunes, Martes, ... Domingo): ").capitalize() # Primera letra en mayúscula
estado_tiempo = input("¿Cómo está el tiempo? (Soleado, Nublado, Lluvioso): ").capitalize()

# Creamos una tupla para facilitar la coincidencia de patrones
info_dia = (dia, estado_tiempo)

match info_dia:
    case ('Sabado' | 'Domingo', 'Soleado'):
        # Coincide si es Sábado O Domingo Y Soleado
        print(f"¡Es {dia} y hace Sol! ¡Perfecto para salir!")
    case ('Sabado' | 'Domingo', _):
        # Coincide si es Sábado O Domingo, sin importar el tiempo (el '_' captura cualquier valor)
        print(f"Es {dia}, fin de semana, pero el tiempo es {estado_tiempo}.")
    case (_, 'Lluvioso'):
        # Coincide cualquier día que sea Lluvioso
        print(f"Es {dia} y está Lloviendo. Mejor quedarse en casa.")
    case ('Lunes', 'Nublado'):
        # Coincide exactamente si es Lunes y Nublado
        print("Es Lunes y está nublado... ¡A empezar la semana con energía!")
    case _:
        # Caso por defecto si no coincide con ninguno de los anteriores
        print(f"Información para el día {dia} con tiempo {estado_tiempo} no está en nuestros patrones definidos.")
````

## Diferencias y Uso de `if/elif/else` vs. `match case`

Ambas sentencias sirven para tomar decisiones, pero son óptimas para diferentes escenarios.

### `if/elif/else` (Más flexible y general)

*   **Cuándo usarlo**: Ideal cuando las condiciones son más complejas y variadas.
    *   Comparaciones con operadores lógicos (`and`, `or`, `not`).
    *   Comparaciones de rangos (`numero > 10 and numero < 20`).
    *   Verificación de múltiples variables en una sola condición.
    *   Cuando las condiciones no se basan únicamente en el valor de una sola expresión.
*   **Ventajas**: Extremadamente versátil y puede manejar casi cualquier tipo de lógica condicional.
*   **Desventajas**: Puede volverse verboso y difícil de leer si tienes muchas `elif` que solo comparan el valor exacto de una misma variable.

### `match case` (Más conciso para coincidencia de patrones)

*   **Cuándo usarlo**: Ideal cuando necesitas comparar el valor de una **única expresión** contra varios **patrones específicos**.
    *   Menús basados en opciones de usuario.
    *   Procesamiento de comandos o mensajes.
    *   Desestructuración de datos (como tuplas o diccionarios) y asignación a variables.
    *   Cuando las condiciones son sobre el **valor o la estructura** de una variable.
*   **Ventajas**: Código más limpio y legible para múltiples comparaciones de valores exactos o patrones de datos. Permite la coincidencia de patrones avanzada.
*   **Desventajas**: No es adecuado para rangos o condiciones lógicas complejas que involucran varias variables o cálculos.

**En resumen:**
*   Usa `if/elif/else` para lógica condicional general y compleja.
*   Usa `match case` para selección basada en valores específicos o la estructura de datos de una sola expresión.

````python
# Ejercicio 6: Elegir entre if/elif/else y match case para un sistema de puntos

# Imaginemos un sistema de recompensas basado en puntos y el nivel del usuario.
# Si un usuario tiene pocos puntos, pero es 'VIP', se le da una recompensa especial.

# Pedimos los puntos y el estado del usuario
puntos = int(input("Ingresa los puntos del usuario: "))
estado_usuario = input("¿Cuál es el estado del usuario? (Normal/VIP): ").capitalize()

print("\n--- Usando if/elif/else para decisiones complejas ---")
# Aquí if/elif/else es mejor porque hay una combinación de condiciones (puntos Y estado)
if puntos >= 1000:
    # Si los puntos son altos, siempre hay una recompensa mayor
    print("¡Recompensa de Oro! Has acumulado muchos puntos.")
elif puntos >= 500:
    # Si los puntos son medianos
    print("Recompensa de Plata: Has alcanzado un buen nivel.")
elif estado_usuario == 'VIP' and puntos < 500:
    # Anidación implícita: Si los puntos son bajos, pero el usuario es VIP
    # Esta condición es muy específica y combina dos variables, ideal para if/elif/else
    print("Recompensa especial VIP: Acceso anticipado a eventos.")
else:
    # Para todos los demás casos (puntos bajos y no VIP)
    print("Recompensa de Bronce: Sigue acumulando puntos.")

print("\n--- Usando match case para un menú de acción simple ---")
# Ahora, si el usuario quiere realizar una acción, un match case es más limpio
accion = input("¿Qué acción deseas realizar? (Ver_Perfil, Ver_Puntos, Salir): ").capitalize()

match accion:
    case 'Ver_perfil':
        print("Mostrando tu perfil de usuario.")
    case 'Ver_puntos':
        print(f"Tus puntos actuales son: {puntos}.")
    case 'Salir':
        print("Cerrando sesión. ¡Hasta pronto!")
    case _:
        print("Acción no reconocida.")
````
