## 1. Definición y Estructura Básica

El bucle `for` se utiliza para **iterar** sobre una secuencia (como una lista, una cadena de texto o un rango de números). A diferencia de otros lenguajes, en Python el `for` es un 'foreach', lo que significa que recorre elementos.

### Estructura:
```python
for variable in secuencia:
    # Bloque de código a repetir
    # variable es la que crea el for, secuencia es la que yo le entrego
```

````python
# Ejemplo 1: El uso de range(inicio, fin, paso)
# range() nos permite controlar exactamente el flujo de la secuencia:
# 1. inicio: El primer número de la secuencia (por defecto 0).
# 2. fin: El límite donde termina (NO se incluye este número).
# 3. paso: De cuánto en cuánto salta (por defecto 1).

print("Contando del 2 al 10, de 2 en 2:")
# inicio=2, fin=11 (para incluir el 10), paso=2
for i in range(2, 11, 2):
    print(f"Valor actual de i: {i}")

# ¿Qué es 'i'?
# 'i' es una variable que el bucle crea automáticamente.
# En cada vuelta (iteración), toma el valor del siguiente elemento generado por range.
````

## 2. Diferencia con el Bucle `while`

*   **`for`**: Se usa cuando sabemos (o podemos calcular) cuántas veces se va a repetir el ciclo.
*   **`while`**: Se usa mientras una condición sea verdadera. Es ideal cuando no sabemos exactamente cuándo terminará.

````python
# Comparación: Imprimir números del 1 al 3

print("Con FOR:")
for j in range(1, 4):
    print(j)

print("\nCon WHILE:")
contador = 1
while contador <= 3:
    print(contador)
    contador += 1 # Importante: ¡en el while debemos actualizar nosotros la variable!
````

````python
# OTRO EJEMPLO DE DIFERENCIA:
# Escenario: Queremos pedir un número positivo.

print("--- Ejemplo con WHILE (Validación) ---")
# Usamos while porque no sabemos cuántas veces se equivocará el usuario
numero = -1
while numero < 0:
    # Simulamos entrada de datos válida tras un intento
    print("Intento de ingreso...")
    numero = 5
print(f"Número aceptado: {numero}")

print("\n--- Ejemplo con FOR (Recorrido fijo) ---")
# Usamos for porque sabemos exactamente que queremos 3 avisos
for intento in range(1, 4):
    print(f"Aviso de seguridad {intento} de 3")
````

### 2.1 Ejercicios Resueltos por Niveles

Aquí veremos cómo el bucle `for` puede resolver problemas desde lo más simple hasta lógica más compleja.

````python
# NIVEL BÁSICO: Imprimir los múltiplos de 5 hasta el 25
print("--- Nivel Básico ---")
for i in range(5, 26, 5):
    print(f"Múltiplo encontrado: {i}")
````

````python
# NIVEL INTERMEDIO: Contar cuántas letras 'a' tiene una frase
frase = "Aprendiendo programación en la Santo Tomas"
contador_a = 0

for letra in frase.lower():
    if letra == 'a':
        contador_a += 1

print("\n--- Nivel Intermedio ---")
print(f"La frase tiene {contador_a} letras 'a'.")
````

````python
# NIVEL AVANZADO: Encontrar números primos en un rango
print("\n--- Nivel Avanzado ---")
primos = []

for num in range(2, 21):
    es_primo = True
    for i in range(2, num):
        if num % i == 0:
            es_primo = False
            break
    if es_primo:
        primos.append(num)

print(f"Números primos entre 2 y 20: {primos}")
````

## 3. Combinando todo: Listas, Acumuladores e `if/match`

En programación real, usamos el `for` para llenar listas, sumar valores (acumuladores) y tomar decisiones dentro del bucle.

````python
# Ejemplo Complejo: Clasificador de Números

numeros_ingresados = []
suma_pares = 0

# Usamos for para repetir una acción 5 veces
for vez in range(1, 6):
    # Solicitamos datos (usamos un ejemplo fijo para que corra en el notebook)
    num = vez * 2  # Simulamos entrada de datos

    # Llenar una lista
    numeros_ingresados.append(num)

    # Condicional IF para el acumulador
    if num % 2 == 0:
        suma_pares += num

    # Estructura MATCH (Python 3.10+) para mensajes especiales
    match num:
        case 10:
            print("¡Se alcanzó la decena!")
        case _:
            pass

print(f"Lista de números: {numeros_ingresados}")
print(f"Suma de los pares: {suma_pares}")
````

## 4. Ejercicios Resueltos (Para estudio)

**Problema:** Crear una lista con los cuadrados de los números del 1 al 5 usando un bucle.

````python
cuadrados = []
for n in range(1, 6):
    resultado = n ** 2
    cuadrados.append(resultado) # Agregamos a la lista
    print(f"El cuadrado de {n} es {resultado}")

print(f"Lista final: {cuadrados}")
````
