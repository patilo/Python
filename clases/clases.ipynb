## 1. Nivel Básico: Definición y Funciones (Métodos)

Una clase es un 'molde' para crear objetos. Los métodos son funciones definidas dentro de la clase que operan sobre los datos del objeto (`self`).

```python
class Mascota:
    def __init__(self, nombre, especie):
        # El constructor inicializa los atributos básicos de la mascota
        self.nombre = nombre
        self.especie = especie

    def saludar(self):
        # Método que devuelve un saludo usando los atributos de la instancia
        return f"Hola, soy {self.nombre} y soy un {self.especie}."

# Ejemplo de uso
mi_perro = Mascota("Rex", "Perro")
print(mi_perro.saludar())
```

### 1.1 Ejemplo Interactivo con `input()`

Podemos pedirle al usuario que ingrese los datos para personalizar nuestro objeto.

```python
class MascotaInteractiva:
    def __init__(self, nombre, especie):
        # Asigna el nombre y especie recibidos por parámetros
        self.nombre = nombre
        self.especie = especie

    def saludar(self):
        # Retorna una cadena formateada con los datos del objeto
        return f"¡Guau/Miau! Soy {self.nombre} y soy un {self.especie}."

# Pedimos datos al usuario por consola
print("--- Registro de Mascota ---")
nombre_usuario = input("¿Cómo se llama tu mascota? ")
especie_usuario = input("¿Qué animal es? ")

# Creamos el objeto pasando las variables capturadas
mi_nueva_mascota = MascotaInteractiva(nombre_usuario, especie_usuario)
print(mi_nueva_mascota.saludar())
```

### 1.2 Registro Dinámico con `input()` y Ciclos

Este ejemplo permite al usuario decidir cuántos elementos quiere agregar a la clase usando un bucle.

```python
class RegistroEstudiantes:
    def __init__(self):
        # Inicializa una lista vacía para almacenar diccionarios de estudiantes
        self.lista_estudiantes = []

    def agregar(self, nombre, nota):
        # Añade un nuevo estudiante a la lista interna
        self.lista_estudiantes.append({"nombre": nombre, "nota": nota})

    def mostrar_reporte(self):
        # Recorre la lista y muestra los datos de cada estudiante
        print("\n--- Reporte Final ---")
        for est in self.lista_estudiantes:
            print(f"Estudiante: {est['nombre']} | Nota: {est['nota']}")

# Instancia de la clase
registro = RegistroEstudiantes()

# Bucle infinito que se rompe cuando el usuario escribe 'salir'
while True:
    nombre = input("Ingrese nombre del estudiante (o 'salir' para terminar): ")
    if nombre.lower() == 'salir':
        break

    try:
        # Convertimos la entrada a float para asegurar que sea numérica
        nota = float(input(f"Ingrese la nota de {nombre}: "))
        registro.agregar(nombre, nota)
    except ValueError:
        # Manejo de error si el usuario ingresa texto en lugar de un número
        print("Error: La nota debe ser un número. Inténtalo de nuevo.")

registro.mostrar_reporte()
```

## 2. Nivel Intermedio: Clases con Ciclos

Podemos usar ciclos para procesar colecciones de datos dentro de una clase.

```python
class CarritoDeCompras:
    def __init__(self):
        self.productos = []

    def agregar_producto(self, producto, precio):
        self.productos.append({"nombre": producto, "precio": precio})

    def calcular_total(self):
        total = 0
        # Usamos un ciclo for para recorrer la lista
        for item in self.productos:
            total += item["precio"]
        return total

# Ejemplo
carrito = CarritoDeCompras()
carrito.agregar_producto("Libro", 25.50)
carrito.agregar_producto("Café", 5.00)
print(f"Total en el carrito: ${carrito.calcular_total()}")
```

### 2.1 Ejemplo Intermedio Interactivo

En este ejemplo, permitimos que el usuario llene el carrito de compras manualmente usando un ciclo y entradas de teclado.

```python
class CarritoInteractivo:
    def __init__(self):
        #Lsista para guardar los productos comprados
        self.items = []

    def agregar(self, nombre, precio):
        #Inserta un producto y su precio al carrito
        self.items.append({"nombre": nombre, "precio": precio})

    def mostrar_resumen(self):
        #Calcula el total y muestra cada item con formato de dos decimales
        print("\n--- Resumen de Compra ---")
        total = 0
        for i in self.items:
            print(f"- {i['nombre']}: ${i['precio']:.2f}")
            total += i['precio']
        print(f"Total a pagar: ${total:.2f}")

mi_carrito = CarritoInteractivo()

print("--- Bienvenido a la Tienda ---")
while True:
    prod = input("Producto a agregar (o 'pagar' para finalizar): ")
    if prod.lower() == 'pagar':
        break

    try:
        #Intentamos capturar el precio y validarlo
        costo = float(input(f"Precio de {prod}: "))
        mi_carrito.agregar(prod, costo)
    except ValueError:
        #Si la conversion a float falla, informamos al usuario
        print("Precio invalido. Intente de nuevo.")

mi_carrito.mostrar_resumen()
```

## 3. Nivel Avanzado: Control de Errores con `try-except`

En aplicaciones reales, debemos validar los datos usando bloques `try` para evitar que el programa se detenga por errores inesperados.

```python
class CalculadoraInteractiva:
    def dividir(self, a, b):
        """
        Realiza una division manejando posibles errores criticos.
        """
        try:
            # Intentamos realizar la operacion
            resultado = a / b
        except ZeroDivisionError:
            # Se ejecuta si el divisor es 0
            return "Error: No se puede dividir entre cero."
        except TypeError:
            # Se ejecuta si los tipos de datos no permiten la operacion
            return "Error: Operación no valida para estos tipos de datos."
        else:
            # Se ejecuta solo si no hubo ninguna excepcion
            return f"Resultado: {resultado:.2f}"
        finally:
            # Se ejecuta siempre, haya error o no
            print("calcuilo procesado.")

# Instanciamos la clase
calc = CalculadoraInteractiva()

print("--- Calculadora Avanzada ---")
try:
    # Capturamos las entradas del usuario
    num1 = float(input("Ingrese el primer numero (dividendo): "))
    num2 = float(input("Ingrese el segundo numero (divisor): "))

    # Llamamos al método de la clase
    mensaje = calc.dividir(num1, num2)
    print(mensaje)
except ValueError:
    # Manejo de error si el usuario ingresa texto en lugar de numeros en el input
    print("Error: Debes ingresar valores numericos validos en la consola.")
```

```python
class SistemaInventario:
    def __init__(self):
        # Diccionario para almacenar: {ID: {'nombre': str, 'precio': float}}
        self.productos = {}

    def registrar(self, id_prod, nombre, precio):
        # Agrega o actualiza un producto en el diccionario
        self.productos[id_prod] = {"nombre": nombre, "precio": precio}

    def buscar(self, id_prod):
        try:
            # Intentamos acceder a la llave del diccionario
            item = self.productos[id_prod]
            return f"Producto: {item['nombre']} | Precio: ${item['precio']:.2f}"
        except KeyError:
            # Se lanza si el ID no existe en el diccionario
            return f"Error: El ID '{id_prod}' no se encuentra en el sistema."

# Lógica del programa
inventario = SistemaInventario()
print("--- Registro de Inventario (Nivel 3) ---")

# Ciclo para poblar el diccionario
while True:
    opcion = input("\n¿Desea (R)egistrar, (B)uscar o (S)alir?: ").upper()

    if opcion == 'S':
        break

    elif opcion == 'R':
        try:
            uid = input("Id unico del producto: ")
            nom = input("Nombre del producto: ")
            pre = float(input("Precio unitario: "))
            inventario.registrar(uid, nom, pre)
            print("¡Producto guardado exitosamente!")
        except ValueError:
            print("Error: El precio debe ser numero decimal.")

    elif opcion == 'B':
        uid_busqueda = input("Ingrese el ID a buscar: ")
        print(inventario.buscar(uid_busqueda))

    else:
        print("opcion no valida.")

print("Sistema cerrado.")
```
